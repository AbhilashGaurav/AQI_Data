```python
# https://app.cpcbccr.com/AQI_India/ to downlaod the excel file

# import pandas lib as pd
import pandas as pd

# read by default 1st sheet of an excel file
dataframe1 = pd.read_excel('/content/sample_data/1 march.xlsx')

# dropping the unwanted columns

dataframe1.drop([0,1,2],axis=0,inplace=True)

# changing the column names
dataframe1.rename(columns = {'Central Pollution Control Board':'Sr._No.'}, inplace = True)
dataframe1.rename(columns = {'Unnamed: 1':'State'}, inplace = True)
dataframe1.rename(columns = {'Unnamed: 2':'City'}, inplace = True)
dataframe1.rename(columns = {'Unnamed: 3':'Station_Name'}, inplace = True)
dataframe1.rename(columns = {'Unnamed: 4':'Current_AQI_value'}, inplace = True)
```

```python
# resetting the index of the data frame
dataframe_indi = dataframe1.reset_index()

# droping the coulmn 0 of the dataframe
dataframe_indi.pop(dataframe_indi.columns[0])
```

```python
df = dataframe_indi.fillna(method='ffill') # this is for filling the vacc
df.head(20)
```

```python
import plotly.graph_objects as go

def aqi_graph(value, graph_color,condn):
  # print(graph_color)
  fig = go.Figure(go.Indicator(
      value = value,
      mode = "gauge+number",
      domain = {'x': [0, 1], 'y': [0, 1]},
      title = {'text': "AQI is {}".format(condn), 'font': {'size': 26}},
      # delta = {'reference': 400, 'increasing': {'color': "RebeccaPurple"}},
      gauge = {
          'axis': {'range': [None, 500]},
          'bar': {'color': graph_color},# this is the bar color
          'bgcolor': "white",
          'borderwidth': 2,
          'bordercolor': "gray",
          # 'steps': [
          #     {'range': [0, 250], 'color': 'cyan'},
          #     {'range': [250, 400], 'color': 'royalblue'}],
          'threshold': {
              'line': {'color': "red", 'width': 4},
              'thickness': 0.75,
              'value': 401}}))

  fig.show()
```

```python
value = 202
color = ''
condn = ''
if (value>=0 and value<=50):
  color = 'Green'
  condn = 'Good'
elif (value>=51 and value<=100):
  color = '#90EE90'
  condn = 'Satisfactory'
elif (value>=101 and value<=200):
  color = 'yellow'
  condn = 'Moderate'
  
elif (value>=201 and value<=300):
  color = 'orange'
  condn = 'Poor'
elif (value>=301 and value<=400):
  condn = 'Very Poor'
  color = 'red'
elif (value>=401 and value<=500):
  color = '#8B0000' # hex code for dark red
  condn = 'Severe'
elif(value>500):
  color = '#8B0000' # hex code for dark red
  condn = 'Severe'
else:
  color = 'green'
  condn = 'Good'


```

```python
aqi_graph(value,color,condn)
```
